```python
title = []
submit = []
review = []
department = []
context = []
for i in range(206,255):
    txt_parsing = driver.find_element_by_xpath('//*[@id="nttViewForm"]/div[1]/table/tbody')
    txt_parsing2 = txt_parsing.text.split('\n')
    txt_parsing3 = []
    for k in txt_parsing2:
        txt_parsing3.append(k.strip())
    title.append(txt_parsing3[0])
    submit.append(txt_parsing3[2].split(' ')[1])
    review.append(txt_parsing3[3].split(' ')[1])
    try:
        department.append(txt_parsing3[1].split(' ')[1])
    except:
        department.append(txt_parsing3[4].split(' ')[1])
    string = ''
    try:
        aaa = txt_parsing3.index('첨부파일')
    except:
        aaa = txt_parsing3.index('다음글')
        
    for j in txt_parsing3[4:aaa]:
        string = string+'|'+j
    context.append(string)
    time.sleep(2)
    if i == 1:
        driver.find_element_by_xpath('//*[@id="sub_content"]/div[2]/ul/li/a').click()
    else:
        try:
            driver.find_element_by_xpath('//*[@id="sub_content"]/div[2]/ul/li[1]/a').click()
        except:
            try:
                driver.find_element_by_xpath('//*[@id="nttViewForm"]/div/table/tbody/tr[5]/td/span/ul/li[1]/a').click()
            except:
                driver.find_element_by_xpath('//*[@id="nttViewForm"]/div/table/tbody/tr[7]/td/ul/li[1]/a').click()
    time.sleep(2)
    print(i)
```

```python
b = driver.find_element_by_xpath('//*[@id="right"]/div[2]/div[1]/table')
c = b.find_elements_by_tag_name('tr')
len(c)
```

```python
for count in range(0,35):
    b = driver.find_element_by_xpath('//*[@id="right"]/div[2]/div[1]/table/tbody')
    c = b.find_elements_by_tag_name('tr')
    hh = []
    for i in c:
        hh.extend(i.find_elements_by_tag_name('td'))
        hh.extend(i.find_elements_by_tag_name('th'))
    lst3 = []
    for i in hh:
        lst3.append(i.text)
    if count == 0:
        new_array = np.array(lst3).reshape(5,9)
        stacking = new_array
    else:
        new_array = np.array(lst3).reshape(10,9)
        stacking = np.vstack((stacking,new_array))
    driver.find_element_by_xpath('//*[@id="paging_bar"]/a[2]/img').click()
```

```python
abc = pd.DataFrame(stacking)
abc.head()
```

```python
abc.to_excel('erume_cheiup_content.xlsx')
```

```python
lst = []
for counts in range(0,34):
    if counts ==0:
        numnum = 5
    else:
        numnum = 10
    for num in range(numnum,0,-1):
        driver.find_element_by_xpath('//*[@id="right"]/div[2]/div[1]/table/tbody/tr['+str(num)+']/th[2]/a').click()
        koko = driver.find_element_by_xpath('//*[@id="right"]').text
        if len(koko.split('\n')) >16:
            context = koko.split('\n')[15:][:-1]
            stack_strings = ''
            for i in context:
                stack_strings += i + '|'
            lst.append(stack_strings)
            driver.find_element_by_xpath('//*[@id="right"]/div[2]/div[2]/a').click()
        else:
            lst.append('no context')
            driver.find_element_by_xpath('//*[@id="right"]/div[2]/div[2]/a').click()
    
    driver.find_element_by_xpath('//*[@id="paging_bar"]/a[2]/img').click()
```

```python
pd.DataFrame(lst).to_excel('erume_cheiup_content.xlsx')
```
