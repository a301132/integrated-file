## 정적 웹크롤링

### 웹 페이지를 열었을 때 이 웹 페이지에 있는 것들만 추출하는 방식입니다.

![ㅁㄴㅇ](web3.png)

#### 처음으로 이 웹페이지에 있는 모든 코드들을 불러와 보겠습니다.

```python
'''패키지를 호출합니다.'''
import requests
'''위의 웹사이트 주소를 가져옵니다.'''
URL = "https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=102&oid=437&aid=0000262099"
'''URL 해당하는 소스코드를 모두 가져옵니다.'''
response = requests.get(URL,headers={'User-Agent':'Mozilla/5.0'})
'''headers부분이 없이 해도 긁어올 수 있는 웹페이지들이 있습니다.'''
'''좋은 웹사이트일 경우 crawling을 막습니다. 네이버일 경우 막고 있어서, 우회를 하여 데이터를 크롤링합니다.'''
```

```python
'''print 함수를 통해 가져온 소스코드를 문자열(str)로 변환 후 확인합니다.'''
print(response.text)
```

```python
'''데이터를 파싱하기 위해 필요한 패키지를 호출합니다.'''
from bs4 import BeautifulSoup
'''response.text를 다른 객체에 저장합니다.'''
code = response.text
bs_code = BeautifulSoup(code,'html.parser')
print(bs_code)
```

#### print한 형태를 보게 되면 아무런 차이가 없는 것 같지만 type()함수를 사용하여 차이점을 본다.

```python
type(code), type(bs_code)
```

#### code는 문자열(str)타입이고, bs_code는 bs.BeautifulSoup입니다.
#### 이렇게 type을 변경하는 이유는 데이터의 추출을 용이하게하고 BeautifulsSoup에서 지원하는 함수들을 사용하기 위해 타입을 변경합니다.

#### 우리가 파싱한 코드에서 "정치부회의 #청와대 발제"라는 부분을 추출하고 싶다.
#### 그러기 위해서는 파이썬에 어떠한 것을 추적해서 그 값을 찾아라고 코드를 작성해주어야 한다.
#### Copy부분을 보게되면 element, outerHTML, selector, JS path, style, XPath를 복사할 수 있습니다.
#### 이것들이 무엇을 의미하는지 아주 쉽게 설명을 하게 되면 "정치부회의 #청와대 발제"라는 것을 추적할 수 있는 경로를 얘기해줍니다.
#### 즉, 우리는 저 속성들을 가지고 python에게 이 경로에 있는 값을 추적하라고 지시할 수 있습니다.
#### element, selector를 통해 찾아봅니다.

#### 일단 element와 selector를 카피합니다.

```python
element : <strong class="media_end_summary">정치부회의 #청와대 발제</strong>
selector : #articleBodyContents > strong
```

#### element 요소를 보게되면 우리가 추출해야 되는 값들이 포함되어 나오게 된다. 그러므로

```python
'''element에 strong(tag)이라는 코드로 갇혀있는 데이터를 모두 불러올수 있습니다.'''
bs_code.find_all('strong')
```

```python
'''class 속성을 통해 데이터를 불러옵니다.'''
bs_code.find(class_='media_end_summary')
```

```python
'''selector 속성을 통해 데이터를 불러옵니다.'''
bs_code.select_one('#articleBodyContents > strong')
```

#### 웹페이지에서 추출한 부분을 가져왔습니다. 이제는 "정치부회의 #청와대 발제"부분만 추출해보겠습니다.

```python
'''tag에서는 6번째해당하는 부분을 가져오겠습니다.'''
'''파이썬은 0부터 순서를 세기 때문에 5라고 입력해줍니다.'''
print(bs_code.find_all('strong')[5].text)
print(bs_code.find(class_='media_end_summary').text)
print(bs_code.select_one('#articleBodyContents > strong').text)
```

```python
print(bs_code.find_all(class_='media_end_summary'),',', bs_code.find(class_='media_end_summary'))
```

#### 위의 출력값을 보게 되면 왼쪽은 list(대가로)로 감싸져있는 것을 확인할 수 있습니다.

```python
'''Error가 나오는 것을 확인할 수 있습니다.'''
bs_code.find_all(class_='media_end_summary').text
```

```python
'''두번째 경우에는 바로되었는데 두번째의 경우 에러가 나왔습니다.'''
'''그렇다면 저 대가로가 문제인 것 같습니다.'''
'''대가로(list)를 벗기고 다시 돌려보겠습니다.'''
print(bs_code.find_all(class_='media_end_summary')[0])
print(bs_code.find_all(class_='media_end_summary')[0].text)
'''어떠한 객체가 들어감에 따라 인식을 할 수 없고 인식을 하는지를 알아봤습니다.'''
'''이것은 단기간에 학습을 통해 알아가는 것이아니라 어떠한 객체를 사용해야되는지 Error에 대한 이해를 하고 풀어가는 과정을 반복적으로 학습하면 어떻게 해결할 수 있는지 알 것입니다.'''
```
